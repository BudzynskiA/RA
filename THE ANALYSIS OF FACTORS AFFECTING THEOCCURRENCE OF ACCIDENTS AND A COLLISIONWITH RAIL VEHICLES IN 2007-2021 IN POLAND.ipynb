```python
# Import Pandas Library
import pandas as pd
# Import csv file with data
train_year = pd.read_csv('TRAIN_YEAR.csv',delimiter=';')
```

```python
#Look how data looks like
train_year
```

```python
train_year['VICTIMS'] = train_year['DEAD'] + train_year['HARD INJURED'] + train_year['SLIGHTLY INJURED']
train_year
```

```python
# Import MatplotLib Library
import matplotlib.pyplot as plt
# Setting the figure size
f=plt.figure(figsize=(10,5))
# Declaring the points for first line plot
X1 = train_year['YEAR'] 
Y1 = train_year['INCIDENTS']
# Plotting the first plot
plt.plot(X1, Y1, label = "INCIDENTS")

# Declaring the points for second line plot
X2 = train_year['YEAR'] 
Y2 = train_year['DEAD']
# Plotting the second plot
plt.plot(X2, Y2, label = "DEAD")

# Declaring the points for third line plot
X3 = train_year['YEAR'] 
Y3 = train_year['HARD INJURED']
# Plotting the second plot
plt.plot(X3, Y3, label = "HARD INJURED")

# Declaring the points for fourth line plot
X4 = train_year['YEAR'] 
Y4 = train_year['SLIGHTLY INJURED']
# Plotting the second plot
plt.plot(X4, Y4, label = "SLIGHTLY INJURED")

# Declaring the points for fifth line plot
X5 = train_year['YEAR'] 
Y5 = train_year['VICTIMS']
# Plotting the second plot
plt.plot(X5, Y5, label = "VICTIMS")

# Labeling the X-axis
plt.xlabel('YEAR')
# Labeling the Y-axis
plt.ylabel('QUANTITY')
# Give a title to the graph
plt.title('Annual variability of incidents with trains')
# Show a legend on the plot
plt.legend()
# Show a plot
plt.show()
```

```python
ctrain_year = train_year.corr()
```

```python
crtrain_year = ctrain_year.round(2)
```

```python
import seaborn as sns
sns.heatmap(crtrain_year, annot=True)
```

```python
ptrain_year = train_year.pct_change()
```

```python
ptrain_year
```

```python
ptrain_year.agg("mean")
```

```python
# Import csv file with data
train_month = pd.read_csv('TRAIN_MONTH.csv',delimiter=';')
```

```python
# Setting the figure size
f=plt.figure(figsize=(15,5))
# Declaring the points for first line plot
X1 = train_month['MONTH'] 
Y1 = train_month['QUANTITY INCIDENTS']
# Declaring the points for second line plot
# Plotting the first plot
plt.plot(X1, Y1, label = "QUANTITY INCIDENTS")
# Plotting the first plot
# Labeling the X-axis
plt.xlabel('MONTH')
# Labeling the Y-axis
plt.ylabel('QUANTITY INCIDENTS')
```

```python
train_month = train_month.sort_index(ascending=False)
ax = train_month.plot.barh(x='MONTH', y='QUANTITY INCIDENTS', rot=0,figsize=(13,5))
y = train_month['QUANTITY INCIDENTS']
for index, value in enumerate(y):
    plt.text(value, index, str(value))
```

```python
train_month['QUANTITY INCIDENTS'].agg(["mean","median","max","min","std"])
```
